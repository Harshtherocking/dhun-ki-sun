```python
import os
from pydub import AudioSegment
import torchaudio
import torchaudio.transforms as T
import torchaudio.functional as F
import matplotlib.pyplot as plt

# Define the file path
file_path = r"/content/0_20250302_174503_part003.mp3"

# Check if the file exists
if os.path.exists(file_path):
    print("File found!")

    # Load the MP3 file
    audio = AudioSegment.from_mp3(file_path)

    # Export as WAV
    wav_file_path = file_path.replace(".mp3", ".wav")
    audio.export(wav_file_path, format="wav")

    # Load the WAV file using torchaudio
    waveform, sample_rate = torchaudio.load(wav_file_path)

    # Print the shape of the waveform and sample rate
    print(f"Waveform shape: {waveform.shape}, Sample Rate: {sample_rate}")

    # Resample the audio to 16 kHz (optional)
    new_sample_rate = 16000
    if sample_rate != new_sample_rate:
        resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=new_sample_rate)
        waveform = resampler(waveform)
        sample_rate = new_sample_rate
        print(f"Resampled to {new_sample_rate} Hz")

    # Apply a lowpass filter
    cutoff_freq = 300  # Cutoff frequency in Hz
    waveform_filtered = F.lowpass_biquad(waveform, sample_rate, cutoff_freq)

    # Change speed (optional)
    speed_change = T.Resample(orig_freq=sample_rate, new_freq=int(sample_rate * 0.8))
    waveform_speed_changed = speed_change(waveform_filtered)

    # Create a Mel spectrogram
    mel_spectrogram = T.MelSpectrogram(sample_rate=sample_rate)(waveform_speed_changed)

    # Plot the processed waveform
    plt.figure(figsize=(12, 4))
    plt.plot(waveform_speed_changed.t().numpy()[0])  # Plot the first channel
    plt.title('Processed Waveform')
    plt.xlabel('Time (samples)')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.show()

    # Plot the Mel spectrogram
    plt.figure(figsize=(12, 4))
    plt.imshow(mel_spectrogram.log2()[0, :, :].detach().numpy(), cmap='viridis', aspect='auto')
    plt.title('Processed Mel Spectrogram')
    plt.xlabel('Time (frames)')
    plt.ylabel('Mel Frequency (bins)')
    plt.colorbar(format='%+2.0f dB')
    plt.show()
else:
    print("File not found. Please check the file path.")
```

```python
import torch
import matplotlib.pyplot as plt
import torchaudio
import noisereduce as nr
import os
import numpy as np

# Function to plot waveform
def plot_waveform(waveform, sample_rate, title="Waveform"):
    # Ensure the waveform is in the right format for plotting
    if isinstance(waveform, np.ndarray):
        if waveform.ndim == 1:
            # Single channel numpy array
            plt.figure(figsize=(12, 4))
            plt.plot(waveform)
        else:
            # Multi-channel numpy array
            plt.figure(figsize=(12, 4))
            plt.plot(waveform.T)
    else:
        # Handle torch tensor
        if waveform.ndim == 1:
            plt.figure(figsize=(12, 4))
            plt.plot(waveform.numpy())
        else:
            plt.figure(figsize=(12, 4))
            plt.plot(waveform.t().numpy())

    plt.title(title)
    plt.xlabel("Samples")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()

# Function to play audio
def play_audio(waveform, sample_rate):
    # Convert numpy to tensor if needed
    if isinstance(waveform, np.ndarray):
        if waveform.ndim == 1:
            # Convert 1D numpy array to 2D tensor [1, samples]
            waveform = torch.tensor(waveform).unsqueeze(0)
        else:
            # Convert multi-channel numpy array to tensor
            waveform = torch.tensor(waveform)
    else:
        # Make sure tensor has right dimensions [channels, samples]
        if waveform.ndim == 1:
            waveform = waveform.unsqueeze(0)
        elif waveform.ndim == 3:
            waveform = waveform.squeeze(0)

    # Make sure tensor is contiguous in memory
    waveform = waveform.contiguous()

    # Save as temporary WAV file
    torchaudio.save("temp.wav", waveform, sample_rate)
    from IPython.display import Audio
    return Audio("temp.wav")

# Use your existing audio file
file_path = "/content/0_20250302_174503_part001.mp3"  # Adjust to your file path

if os.path.exists(file_path):
    print(f"Loading file: {file_path}")

    # Load audio file
    # First convert to WAV for torchaudio to work with it
    from pydub import AudioSegment

    # Convert MP3 to WAV
    wav_file_path = file_path.replace(".mp3", ".wav")
    audio = AudioSegment.from_mp3(file_path)
    audio.export(wav_file_path, format="wav")

    # Load with torchaudio
    waveform, sample_rate = torchaudio.load(wav_file_path)
    print(f"Loaded audio: {waveform.shape}, Sample rate: {sample_rate}")

    # If stereo, use only one channel for simplicity in denoising
    if waveform.shape[0] > 1:
        speech = waveform[0].unsqueeze(0)  # Take first channel
    else:
        speech = waveform

    # Create synthetic noise (or use a portion of the audio as noise)
    # Option 1: Synthetic noise
    noise = torch.randn_like(speech) * 0.02

    # Option 2: Use a segment of the audio as noise (if you have a section with just noise)
    # noise_segment = speech[:, :int(speech.shape[1]*0.1)]  # Use first 10% as noise
    # noise = noise_segment.repeat(1, int(speech.shape[1] / noise_segment.shape[1]) + 1)[:, :speech.shape[1]]

    # Calculate POWER (squared L2 norm)
    speech_power = speech.norm(p=2) ** 2
    noise_power = noise.norm(p=2) ** 2

    # Add noise to the original speech
    snr_db = 10  # Example SNR value (10dB)
    snr_linear = 10 ** (snr_db / 10)
    scale = torch.sqrt(speech_power / (snr_linear * noise_power))
    noisy_speech = speech + scale * noise

    # Convert to numpy for noisereduce
    # Make sure it's the right format for noisereduce
    noisy_speech_np = noisy_speech.squeeze().numpy()

    # Remove noise using noisereduce
    print("Applying noise reduction...")
    denoised_speech = nr.reduce_noise(y=noisy_speech_np, sr=sample_rate)

    # Convert denoised speech back to tensor with correct shape [channels, samples]
    denoised_speech_tensor = torch.tensor(denoised_speech).unsqueeze(0)

    # Plot and play original (noise-free) speech
    print("Plotting original speech...")
    plot_waveform(speech, sample_rate, title="Original Speech")
    orig_audio = play_audio(speech, sample_rate)
    display(orig_audio)

    # Plot and play noisy speech
    print("Plotting noisy speech...")
    plot_waveform(noisy_speech, sample_rate, title="Noisy Speech")
    noisy_audio = play_audio(noisy_speech, sample_rate)
    display(noisy_audio)

    # Plot and play denoised speech
    print("Plotting denoised speech...")
    plot_waveform(denoised_speech_tensor, sample_rate, title="Denoised Speech")
    denoised_audio = play_audio(denoised_speech_tensor, sample_rate)
    display(denoised_audio)

else:
    print(f"File not found: {file_path}")
    print("Please check the file path and try again.")
```
